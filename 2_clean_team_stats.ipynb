```python
import pandas as pd
```

## 1. CARGAR LOS CSV

```python
master = pd.read_csv(
    r"Datasets\Jugador_temporada - Basketball Reference\nba_master_dataset_v1.csv"
)

team_summary = pd.read_csv(
    r"Datasets\Team Stats - Kaggle\Team Summaries.csv"
)

team_pg = pd.read_csv(
    r"Datasets\Team Stats - Kaggle\Team Stats Per Game.csv"
)

team_100 = pd.read_csv(
    r"Datasets\Team Stats - Kaggle\Team Stats Per 100 Poss.csv"
)
```

## 2. INSPECCIONAR LAS TABLAS

```python
print(team_summary.shape)
print(team_pg.shape)
print(team_100.shape)

print(team_summary.columns.tolist())

print(team_pg.columns.tolist())

print(team_100.columns.tolist())
```

```python
print(team_summary["season"].min())
print(team_summary["season"].max())

print(team_pg["season"].min())
print(team_pg["season"].max())

print(team_100["season"].min())
print(team_100["season"].max())
```

```python
print(sorted(team_100["season"].unique()))
```

## 3. CONSERVAR ÚNICAMENTE LAS TEMPORADAS 2001-2025

```python
team_summary_clean = team_summary[
    team_summary["season"].between(2001, 2025)
].copy()

team_pg_clean = team_pg[
    team_pg["season"].between(2001, 2025)
].copy()

team_100_clean = team_100[
    team_100["season"].between(2001, 2025)
].copy()
```

```python
def season_format(year):
    return f"{year-1}-{str(year)[-2:]}"
```

```python
for df in [
    team_summary_clean,
    team_pg_clean,
    team_100_clean
]:

    df["Season"] = df["season"].apply(season_format)
```

#### Renombrar columnas

```python
for df in [
    team_summary_clean,
    team_pg_clean,
    team_100_clean
]:

    df.rename(
        columns={
            "abbreviation":"Team"
        },
        inplace=True
    )
```
